```python
# Import modules
import numpy as np
import pandas as pd
import itertools
from copy import deepcopy
from collections import defaultdict
import os, re, ast
import math
from itertools import cycle
import time
```

Read "**sequence.csv**" dataset and put it in a variable T.

```python
with open("sequences.csv", 'r') as f:
    contents = f.read()
T = contents.split("\n")
```

Take each sequence from **T** and put it in the list **sequences**.

```python
sequences = []
for sequence in T:
  sequence = sequence[1:-3]
  sequences.append([ast.literal_eval(str(i)) for i in re.split('(?<=\))\s(?=\()', sequence)])
```

```python
# Print sequences
for sequence in sequences[:3]:
  print(sequence)
```

Take events from each sequence and put them in a list.

```python
events_lists = []
for sequence in sequences:
  temp = []
  for event in sequence:
   temp.append([str(event[-1])])
  events_lists.append(temp)
```

```python
# Print events
for event_row in events_lists[:3]:
  print(event_row)  
```

**C** - Candidate,
**F** - Frequent,
**S** - Sequence,

```python
C = {}
F = {}
S = {}
minsup = 2000
```

```python
# Created Candidate-1 sequence
# Creation of C_1
C_1 = defaultdict(int)
for events_row in events_lists:
  for event in np.unique(events_row):
    if event in C_1:
      C_1[event] += 1
    else:
      C_1[event] = 1
```

```python
C_1
```

```python
# Create Frequet-1 subsequence
# Creation of F_1
F_1 = {}
for item in C_1:
  if C_1[item] >= minsup:
    F_1[item] = C_1[item]
```

```python
F_1 
```

```python
C['C_1'] = C_1
S['S_1'] = list(F_1.keys())
```

Write length-1 sequential patterns to output file. (**output.txt**)

```python
with open('output.txt', 'w') as f:
  for i in S['S_1']:
    f.write("('" + str(i) + "')")
    f.write('\n')
```

A simple function to calculate **length** of an iterable.

```python
def findLen(S):
  count = 0
  for i in S:
    count += 1
  return count
```

```python
# Generate Length-2 Candidates
C_candidate = []
for subset in itertools.permutations(S['S_1'], 2):
  C_candidate.append(subset)
for i in S['S_1']:
  k = list()
  k.append(i)
  k.append(i)
  C_candidate.append(tuple(k))
C['C_2'] = C_candidate
```

# LOOP

We wrote explanation of that part in report in detail.

```python
start = time.time()

with open("output.txt", 'a') as f:

  len = 2
  while findLen(S['S_'+str(len-1)]) !=  0:
    print('INSIDE LOOP', len,'-length Sequence \n')

    if len > 2:
      # CANDIDATE GENERATION
      C['C_'+str(len)] = []
      for i in range(findLen(S['S_'+str(len-1)])):
        for j in range(0, findLen(S['S_'+str(len-1)])):
          if S['S_'+str(len-1)][i][1:] == S['S_'+str(len-1)][j][:-1]:
            C['C_'+str(len)] .append(S['S_'+str(len-1)][i] + S['S_'+str(len-1)][j][len-2:])

      # Pruning
      C_candidate = []
      for i in range(findLen(C['C_'+str(len)] )):
        temp = []
        for subset in itertools.combinations(C['C_'+str(len)][i], len-1):
          temp.append(subset)
        count = 0
        for event in temp:
          if event in S['S_'+str(len-1)]:
            count += 1
        if count == len:
          C_candidate.append(C['C_'+str(len)][i])

    # Support Counting
    C_cand_sup = {}
    for c_event in C_candidate:  
      #print('c_event: ', c_event)
      for events_row in events_lists:
        licycle = cycle(c_event)
        item = next(licycle)
        count=0
        for event in events_row:
          if item in event:
            count += 1
            if count  == len:
              licycle = cycle(c_event)
              item = next(licycle)
              if c_event in C_cand_sup:
                C_cand_sup[c_event] += 1
              else:
                C_cand_sup[c_event] = 1
              #print(c_event, ' is in ', events_row)
              break
            item = next(licycle)
            #print('Item: ', item)  

    # CANDIDATE ELIMINATION
    S['S_'+str(len)] = {}
    for event in C_cand_sup:
      if C_cand_sup[event] >= minsup:
        S['S_'+str(len)][event] = C_cand_sup[event]

    print(S['S_'+str(len)])
    s = ''
    for i in S['S_'+str(len)]:
      s = str(i)
      f.write(str(s))
      f.write('\n')
    S['S_'+str(len)] = list(S['S_'+str(len)].keys())

    len += 1
    
stop = time.time()
time = stop-start
print("Runtime:", time) 
```

# Find **Min** and **Max** timestamp between joined events

```python
d = {}
for sequence in sequences:
  for i in range(findLen(sequence)-1):
    temp = tuple()
    diff = sequence[i+1][0] - sequence[i][0]
    temp += (str(sequence[i][1]),)
    temp += (str(sequence[i+1][1]),)

    if temp in d:
      if diff < d[temp][0]:
        d[temp][0] = diff
      elif diff > d[temp][1]:
        d[temp][1] = diff

    else:
      d[temp] = [diff, diff]
```

This is the function to make output file desired format.

```python
def sequences_with_timestamp(seq_len_2, sequences, filename):
  with open(filename, 'a') as f:
    for temp in sequences:
      i=0
      while i+2 <= findLen(temp):
        subset = temp[i:i+2]
        if subset in seq_len_2.keys():
          f.write('%-1s %-3s %-10s %-2s %-1s ' % ('(', str(subset[0]), str(d[subset]),  str(subset[1]), ')' ))
        i += 1  
      f.write('\n')
```

```python
for i in range(2,findLen(S)+1):
  sequences_with_timestamp(d, S['S_'+str(i)], "output_min_max.txt")
```
